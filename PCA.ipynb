```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('cancer_tumor_data_features.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
#  scaling the data to have equal importance for all features.
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler();
```

```python
scaled_X=scaler.fit_transform(df)
```

```python
scaled_X.mean()  #  means of all feature is near to zero
```

```python
scaled_X.mean(axis=0)
```

```python
scaled_X-scaled_X.mean(axis=0)
```

```python
scaled_X    # both are same since means are zero
```

```python
# calculating covariance and eigen vectors
```

```python
covariance_matrix=np.cov(scaled_X,rowvar=False)
```

```python
# covariance_matrix
```

```python
# Calculating eigen values
```

```python
eigen_values, eigen_vectors = np.linalg.eig(covariance_matrix)
```

```python
# eigen_values
# eigen_vectors
```

```python
eigen_vectors.shape
```

```python
# PCA 30 Feature Space --> 30 Principal Components space --> 2 
# How many PC to choose? N=2
```

# Will use .argsort -> which will return indices acc. to sorted elements

```python
num_components=2
```

```python
sorted_key=np.argsort(eigen_values)[::-1][:num_components]
```

```python
#  extracting top 2 eigen-values & top 2 eigen-vectors
```

```python
eigen_values,eigen_vectors=eigen_values[sorted_key],eigen_vectors[:,sorted_key] #extracting all rows but top 2 columns
```

```python
# Finding Principal components
```

```python
# Original-data --> project --> eigen_vectors
principal_components=np.dot(scaled_X,eigen_vectors) 
```

```python
principal_components
```

```python
# visualizing distribution of these principal components
plt.scatter(principal_components[:,0],principal_components[:,1])
```

```python
# now visualzing acc. to cancer data from sklearn  to get target feature
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer_dictionary=load_breast_cancer()  # returns dictionary
```

```python
cancer_dictionary.keys()
```

```python
cancer_dictionary['target']
```

```python
plt.scatter(principal_components[:,0],principal_components[:,1],c=cancer_dictionary['target'])
```

```python
# NOTICE  it quite nice separated out these into 2 classes with using only 2 features.
```

# Now implementing using sci-kit learn

```python
from sklearn.decomposition import PCA
```

```python
pca_model=PCA(n_components=2)
```

```python
pca_model.fit(scaled_X)  # will set eigen_values & eigen_vectors
```

```python
pca_model.transform(scaled_X)   # doing the projection step (dot product)
```

```python
# doing fit & transform in 1 line
pc_results=pca_model.fit_transform(scaled_X)
```

```python
plt.figure(dpi=200)
plt.scatter(pc_results[:,0],pc_results[:,1],c=cancer_dictionary['target'])
# so we are getting actually the same results
```

```python
#  it showing the directions of maximum variance in the data
pca_model.components_
```

```python
# visulaizing
df_comp=pd.DataFrame(pca_model.components_,index=['PC1','PC2'],columns=df.columns)
```

```python
df_comp
```

```python
plt.figure(figsize=(17,5))
sns.heatmap(df_comp,annot=True)
```

```python
#  finding how much variance each principal components show
pca_model.explained_variance_ratio_
```

```python
#  as you increase the number of components , overall variance ratio will increase
```

```python
np.sum(pca_model.explained_variance_ratio_)    # 2 comp. representing 63%variance of the data
```
